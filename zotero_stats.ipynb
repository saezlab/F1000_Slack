```python
from datetime import datetime
import os
from pyzotero import zotero
from tqdm import tqdm
import pandas as pd

zot = zotero.Zotero(
    os.environ["ZOTERO_LIBRARY_ID"],
    "group",
    os.environ["ZOTERO_API_KEY"],
)

# takes ~ 2.5 min
all_entries = zot.everything(zot.top())

dict_list = []

for entry in tqdm(all_entries):
    dt = datetime.fromisoformat(entry["data"]["dateAdded"].replace("Z", "+00:00"))
    dict_list.append({
        "title": entry["data"]["title"],
        "user_name": entry["meta"]["createdByUser"].get("username"),
        "user_id": entry["meta"]["createdByUser"].get("id"),
        "year": dt.year,
        "month": dt.month,
    })
df = pd.DataFrame(dict_list)
df = df.loc[df["year"]==2025, :]

df.to_csv("zotero_stats_2025.csv")

#df.value_counts("month")
#df.value_counts("user_name")
```
